```python
DB
- 특정 기업이나 조직 또는 개인이 필요에 의해 논리적으로 연관도니 데이터를 모아 일정한 형태로 저장해놓은 것
- 한 조직의 여러 응용 시스템들이 공용르ㅗ 사용하기 위해 통합, 저장한 운영 데이터 집합

DBMS
- DB 관리 프로그램. 입력, 수정, 삭제 등의 기능을 제공함


DB 특징
통합된 데이터
- 데이터를 통합하는 개념
- 각자 사용하던 데이터의 중복을 최소화하여 중복으로 인한 데이터 불일치 현상 제거

저장된 데이터
- 문서로 보관된 데이터가 아닌 디스크, 테이프같은 컴퓨터 저장장치에 저장된 데이터

운영데이터
- 조직의 목적을 위해 사용되는 데이터
- 업무를 위한 검색을 할 목적으로 저장된 데이터

공용 데이터
- 한 사람 또는 한 업무를 위해 사용되는 데이터가 아닌 공동으로 사용되는 데이터


DB는 사이즈의 제한이 없음. 엑셀은 커질 경우 시트를 분리해서 관리 필요함

장점
데이터 중복 최소화 및 공용, 일관성&무결성 유지, 보안 보장, 표준화 용이, 전체 데이터 요구 조정

관계형 데이터베이스 특징
- 2차원 테이블로 표현(row, column)
- 데이터 무결성
- 데이터 처리를 위해 SQL 사용
- 한번에 레코드를 처리하는 것이 아니라 집합을 처리함

기본키/외부키(primary key/foreign key)
- 기본키는 주로 id이며 이는 고유한 값을 가지는 경우에 쓰임
- 참조하여 만들 수 있도록 만들어 놓은 컬럼(키)를 외부키라고 함(연관관계를 가지는 컬럼을 FK로 설정)


SQL(Structured Query Language)이란?
- 관계형 데이터베이스에서 데이터 조작과 데이터 정의를 하기 위해 사용하는 언어
- 사용자는 데이터베이스에 자료를 입력/조회/수정/삭제 하기 위해 사용

특징
- 배우고 사용 쉬움
- 절차적인 언어가 아닌 선언적 언어
- DBMS 종류에 얽매이지 않고 사용 가능

SQL 작업 종류
- 데이터 조회
- 테이블에서 행 삽입, 갱신, 삭제
- 객체 생성, 변경 및 삭제
- 데이터베이스 및 해당 객체에 대한 액세스 제어
- 데이터베이스 일관성 및 무결성 보장

SQL문
1. DML(중요함)
- select : 데이터 검색
- insert : 테이블에서 새 행(row)를 입력
- update : 기존 행 변경
- delete : 행 제거
- merge : 데이터가 테이블에 존재하지 않으면 insert, 존재하면 update 수행
2. DDL
- create : 테이블 등 데이터 구조를 생성
- alter : 객체를 수정할 때 사용
- drop : 객체 제거 시 사용
- rename : 객체 이름을 변경할 때 사용
- truncate : 객체 내의 모든 행을 삭제
3. TCL
- commit : DML문이 변경한 내용을 관리
- rollback : 데이터에 대한 변경내용은 논리적인 트랜잭션으로 그룹화 될 수 있음
4. DCL
- grant, revoke : 오라클 DB 및 해당 구조에 대한 액세스 구너한을 부여하거나 제거
```

데이터 유형
- 테이블 생성시 컬럼을 정의할 때 선언한 데이터 유형은 그 컬럼이 받아들일 수 있는 자료의 유형을 규정
-> 다른 유형이 들어오면 에러 발생
- 오라클에서 제공하는 기본 데이터 타입 종류는 크게 문자형 데이터, 숫자형 데이터, 날짜 데이터로 구분됨
- 데이터 유형과 더불어 지정한 크기(size)도 중요한 기능을 제공
-> 선언 당시에 지정한 데이터의 크기를 넘기는 입력은 에러 발생

데이터 타입
char : 고정 길이의 문자데이터 - 최대 2000바이트, size값 생략시 기본값 1
varchar2 : 가변길이의 문자데이터 - 최대 4000바이트, 사이즈값 생략 불가
clob : long타입 확장 형태, 가변길이 문자데이터 저장 - 최대 4GB
long : 가변 길이 문자 데이터 - 최대 2GB

문자 데이터 타입
char(10) -> happy(공백)(공백)(공백)(공백)(공백)
varchar2(10) -> happy

숫자 데이터 타입
number(prec, scale) : 정수를 포함한 모든 실수를 사용, 38자리 양수 및 음수, precision(전체자릿수), scale(소수점 이하 자릿수): 생략한 경우 적용할 수 있는 값 0 => 최대값 21바이트
binary_float : 32비트 부동소수 (4byte)
binary_double : 64비트 부동소수 (8byte)

날짜 데이터 타입
date : 고정길이 날짜와 시간 데이터, bc 4712년 1월 1일부터 9999년 12월 31일까지 표현 = 7byte, NLS_DATE_FORMAT
timestamp : millisecond까지 표현 가능한 데이터 타입, 0-9까지 가능, 기본 6 = 7~11byte, NLS_TIMESTAMP_FORMAT
timestamp with time zone : 시간대와 함께 날짜와 시간 표현 = 13byte, NLS_TIMESTAMP_TZ_FORMAT
timestamp with local time zone : 저장시 DB시간대 준수, 조회시 클라이언트 시간대로 표현 = 7~11byte, NLS_TIMESTAMP_FORMAT
