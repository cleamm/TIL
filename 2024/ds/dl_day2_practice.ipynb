```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('DL_data/data/house_train.csv')
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df_corr = abs(df.corr().SalePrice).sort_values(ascending=False)
df_corr = df_corr[df_corr>0.5][1:] # saleprice 열 제외
df_corr_index = list(df_corr.index)
df_corr_index
```

```python
# df_corr = df.corr().SalePrice.sort_values(ascending=False)
# df_corr_sort = df_corr[df_corr>0.5][1:]
# df_corr_index = list(df_corr_sort.index)
# df
```

```python
X = df[df_corr_index]
y = df.SalePrice
print(X.shape)
X.head()
```

```python
X.isnull().sum() # 없음
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=0)
```

```python
model = Sequential()
model.add(Dense(64, input_dim = 14, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1)) # activation=linear 생략
model.compile(loss='mse', optimizer='adam')
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=10)
checkpointer = ModelCheckpoint(filepath='housePrice2.h5', monitor='val_loss', verbose=1, save_best_only=True)
```

```python
history = model.fit(X_train, y_train, validation_split=0.25, epochs=10000, batch_size=32, callbacks=[early_stop, checkpointer])
```

```python
df2 = pd.read_csv('DL_data/data/house_test.csv')
df2.head()
```

```python
df2 = pd.get_dummies(df2)
df2.head()
```

```python
df2 = df2.loc[:,df_corr_index]
print(df2.shape)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.fillna(df2.mean(), inplace=True)
```

```python
pred = model.predict(df2)
```
