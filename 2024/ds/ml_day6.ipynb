# DBSCAN의 예시 코드

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# 샘플 데이터 생성
from sklearn.datasets import make_moons
X, _ = make_moons(n_samples=300, noise=0.05)

# DBSCAN 모델 생성 및 학습
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# 클러스터링 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
# 선형회귀: 직선형태를 띄는 모델(회귀는 연속형 값을 예측하는 것임)
# linear regression = 일차함수 -> y = w(회귀계수=가중치)x + b
# 독립변수(x), 종속변수(y)= 어떤 사건의 원인이 되는 변수, 독립변수의 변화에 따른 결과를 나타내는 변수
# 광고비(x), 매출액(y)
# y = w1x1(광고비) + w2x2(교육비) + w3x3(유통비) + ...
# 선형회귀분석에서 주어진 데이터로부터 가장 알맞은(오차의 합이 최소가 되도록 하는) 함수를 구해야 함
# 즉 MSE(평균제곱오차 mean squared error)를 구해야 됨 = 데이터포인트에서 함수 선까지의 거리(차이)의 합계
# 결국 목적은 mse를 최소화하는 함수를 찾는 것이 선형회귀를 구하는 방법
# 선형회귀는 딥러닝 기초의 64페이지부터 확인 가능

# Gradient Descent Algorithm
# 손실 함수를 최소화하는 알고리즘
# 학습률을 높게 설정하면 코스트가 높아지며, 가다가 inf가 나오고 이후에 NaN으로 출력됨(발산)
# 발산하는 경우는 주로 학습률을 잘못 설정한 경우임
```

```python
feature = data[ ['Sepal length','Sepal width']]
feature.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(feature)
```

```python
model.predict(feature) # 실행할 때마다 숫자가 달라짐
```

```python
predict = pd.DataFrame(model.predict(feature))
predict.columns=['predict']
```

```python
r = pd.concat([feature,predict],axis=1)
```

즉 데이터간 차이의 정도가 약했기 때문에 오히려 예측에 방해가 된 것이라고 보면 됨<br>
이런 분류기를 만들거라면 의사결정트리를 사용하거나 svm 등의 분류기를 이용하는 것이 맞음<br>

```python
ks = range(1,10)
inertias = []
```

```python
for k in ks: 
    model = KMeans(n_clusters=k) 
    model.fit(feature) 
    inertias.append(model.inertia_)
```

```python
inertias
```

```python
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

위와 같은 그래프는 '엘보(elbow)그래프'라고 하며,
