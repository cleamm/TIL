```python
# 리빙 라스베가스와 토이 스토리를 비교해서 추천하고 싶을 때 코사인 유사도를 사용한다면, 
# 토이스토리와 라스베가스 영화의 좌표 각도를 비교하는 것임
# 그럼 이 각도는 몇차원의 각도일까? 671차원 벡터의 각도라고 할 수 있음
# 각도가 작을수록 굉장히 닮은 것임 = 점수의 분포도 비슷해짐
# 결론적으로 각도라는 것이 겉으로는 2차원일 수 있는데 좌표를 따졌을 땐 좌표를 구성하는 피처의 종류가 많으므로 피처수만큼 차원이 늘어남
# 유클리드 거리는 알고 있으니 코사인 유사도와 유클리드 유사도를 비교해서 어느쪽이 좋은지 봐야함
```

ml 3일차에 설명한 내용임

유클리드 거리로 계산하면 부족한 요소도 있음

추천영화시스템으로 A(기본) = (20,0) B(사랑) = (0,20) C(돈) = (100, 0) 이렇게 있을 때 거리로 따지면 B라는 것과 연결 짓겠지만 분류로는 C에 가까움. 이러면 유클리드 거리를 쓰는 것이 틀린 방법이 됨.

이에 따라 cos(코사인)유사도 방식으로 구하는 방법을 적용

 ?|딥러닝|머신러닝|나는|
|--|---|--|---|
영화1|0|1|1|
영화2|1|0|1|
영화3|2|2|2|
영화4|1|1|1|

이렇게 있다면 3차원 공간임(피처 개수로 보면 됨)
영화4와 비슷한 영화는? = 영화1,2
영화3과 4는 방향이 완전히 일치함. 다만 크기가 다름
그래서 코사인 유사도를 구할 땐 방향과 크기가 같이 들어가는데 중요한 건 각도임

```python
m1=np.array([0,1,1])
m2=np.array([1,0,1])
m3=np.array([2,2,2])
m4=np.array([1,1,1])
```

```python
def cossim(a,b): # 코사인 유사도 구하는 함수
    return np.dot(a,b) / (np.linalg.norm(a) * np.linalg.norm(b))
print(cossim(m1,m4))
print(cossim(m2,m4))
print(cossim(m3,m4)) # 제일 유사함
```

영화4와 영화4의 유사도는? = 1임.
같은 영화는 유사도를 보는 것이 의미가 없음

```python
# 영화가 입력되면 유사도가 가장 높은 5편의 영화 제목을 출력

import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('movies/movies.csv')
ratings = pd.read_csv('movies/ratings.csv')
print(movies.shape)
print(ratings.shape)
movies.head()
```

```python
movie = input('input movie name')
def top5_cosine_similarity_movies(movie_title):
    df = pd.merge(movies, ratings).drop('timestamp', axis=1)
    rating_title = df.pivot_table('rating', index='title', columns='userId')
    rating_title.fillna(0, inplace=True)
    res = pd.DataFrame(cosine_similarity(rating_title), index=rating_title.index, columns=rating_title.index)
    res_lst = res.drop(movie_title, axis=1).loc[movie_title].sort_values()[-5:]
    print(res_lst)
top5_cosine_similarity_movies(movie)
```
