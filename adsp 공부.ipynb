<h1>데이터 이해</h1>


데이터 : 가공되지 않은 있는 그대로의 객관적 사실
정보 : 데이터로부터 가공된 자료(베스트셀러)

정성적 : 자료 특징을 풀어서 설명 - 문자, 언어(기상특보, 주관식 설문 응답)
정량적 : 자료를 수치화 - 수치, 기호 (온도, 풍속)

정형데이터 : 정보 형태가 정해짐(관계형DB, 엑셀, csv)
<strong>반정형데이터 : 뎅터를 설명하는 메타데이터를 포함(로그, html, xml, json) </strong>
비정형데이터 : 형태가 정해지지 않음(sns, 영상, 음악)

<strong>
암묵지 : 개인에게 습득되고 겉으로 드러나지 않음
형식지: 문서, 메뉴얼 등의 형상화된 지식 
</strong>
공통, 표출, 연결, 내면

DIKW 피라미드

Information(정보) : 데이터를 통해 패턴 인식
지식(knowledge) : 패턴통해 예측
지혜(wisdom) : 창의적 산물

데이터베이스 정의
DB : 일정 구조에 맞게 조직화된 데이터 집합
- 스키마 : DB 구조와 제약조건에 관한 전반적 명세
- 인스턴스 : 정의된 스키마에 따라 저장된 값
DBMS: DB를 관리
관계형 DBMS : mssql, mysql, maria DB,...
객체지향 DBMS : 정보를 객체형태로 처리
<strong>NoSQL DBMS : 비정형 데이터 저장 및 처리(Mongo DB,...)</strong>
SQL : 데이터 베이스에 접근할 수 있는 하부언어

<strong>

결측값과 이상값 검색
EDA(탐색적 자료 분석)
통계 시각화를 통해 파악
('저잔재현')

결측값
- 존재하지 않는 데이터, null/NA로 표시. 의미 있을 수 있음
- 단순대치법: 결측값 가지는 데이터 삭제
예시로 complete cases 함수로 False일 경우(결측값) 제거
- 평균대치법
- 단순확률 대치법(KNN)
- 다중 대치법: 여러번 대치(대치-분석-결합)

이상값
- 극단적인 값. 의미 있을 수도 있음(체중3kg)
- 이상값을 항상 제거하지는 않음
1. ESD: 평균으로부터 표준편차의 3배 넘는 데이터는 이상값으로 판단
2. 사분위수(Q1, Q1.5 이런식임...)

통계분석
통계학 개론
전수조사와 표본조사
- 전수조사: 전체 조사. 시간 비용이 큰 소모
- 표본 조사: 일부만 추출하여 모집단 분석

표본추출 방법
1. 랜덤 추출
2. 계통 추출: 번호 부여하여 일정 간격으로 추출
3. 집락 추출법(아파트가 있으면 1동만 추출하는 방식)
4. 층화 추출법=비례층화 추출법 -> 1학년 1명, 2학년 2명,3학년 3명,...
5. 복원/비복원 추출

자료의 척도
1. 질적 척도- 명목(대학교, 성별), 순서(학년, 순위)
2. 양적 척도- 등간척도(덧셈, 뺄셈만 가능), 비율척도(사칙연산이 가능)

추정

2. 구간추정: 모집단이 특정한 구간(95%, 99%를 가장 많이 사용) = 배달 시간

<strong>가설검정</strong>(무조건 나옴)
- 모집단의 특성에 대한 주장을 가설로 세우고 표본조사로 가설의 채택여부를 판정
1. 귀무가설: 일반적으로 생각하는 가설 = 갑론
2. 대립가설: 귀무가설을 기각하는 가설(귀무가설을 반박) = 을박
3. 유의수준: 귀무가설이 참일 때 기각하는 <strong>1종 오류</strong>를 범할 확률의 허용한계(일반적으로 5%)
4. 유의확률: 귀무가설을 지지하는 정도를 나타낸 확률

정상 범주 내에 들어가지 못하는 영역은 기각역 이라고 함.

ex)
귀무가설 = 차이가 없다 or 동일하다
대립가설 = 양측검정(같지 않다) 혹은 단측검정(~쪽이 크다 or 작다)

1. 귀무가설/대립가설 설정
두 학교의 성적은 '동일하다' -> 귀무가설
2. 양측/단측검정 확인
대립가설의 값이 같지 않다 -> 양측검정
3. 일표본 혹은 이표본 확인
두개의 모집단이면 '이표본', 한개의 모집단이면 '일표본'
4. 귀무가설 기각 혹은 채택
p-value(유의확률): 0.55 > 유의수준(0.05) -> 귀무가설 채택
5. 단일표본, 대응표본, 독립표본 확인
서로 다른 모집단에 대한 평균비교 검정 -> 독립표본


비모수검정: 
1. 모집단에 대한 아무런 정보가 없을 때, 
2. 관측 자료가 특정 분포를 따른다고 가정이 불가할 때,
3.  만-휘트니 U검정, 크러스컬-월리스 검정 등

회귀분석
- 독립변수들이 종속변수에 영향을 미치는지 파악하는 분석방법
- y(종속변수) = wx(독립변수)+b
- 잔차: 계산값과 예측값의 차이.(표본집단 기준)
- 오차: 모집단 기준
- 추정방법
  - <strong>최소제곱법</strong>: 잔차의 제곱합이 최소가 되는 회귀계수와 절편을 구하는 방법
  - R-squared: 총 변동 중 회귀모형의 의해 설명되는 변동이 차지하는 비율(0~1이며 1에 가까울수록 제일 좋고 0은 안좋음)


회귀분석 가정
1. 선형성: 종속변수와 독립변수는 선형관계
2. 등분산성: 잔차의 분산이 고르게 분포
3. 정상성(정규성): 잔차가 정규분포의 특징을 가짐
4. 독립성: 독립변수들간 상관관계가 없음
    - 정규성은 Q-Q plot, 샤피로 윌크 검정, 히스토그램, 왜도, 첨도 등으로 확인


회귀분석 종류
1. 단순회귀(선형회귀: LR) 1개의 독립변수 및 종속변수
2. 다중회귀(다중선형회귀: MLR) 2개 이상의 독립변수와 종속변수
3. 다항회귀: 2개 이상의 독립변수와 종속변수가 2차함수 이상의 관계
4. 릿지회귀: L2규제
5. 라쏘회귀: L1규제

<strong>전체 데이터 수 = 자유도 + 1</strong>

회귀 모형의 검정
문제로 예시
ex) 종속변수 height(y)/ 독립변수 age, no_sibilings( x, x)
- p value가 1.658e-09 이므로 0.05보다 작으며 이에 따라 모형이 유의미함
- 위 모형은 다중회귀 모형(독립변수가 2개)이 됨

최적 회귀 방정식 탐색 방법
1. 전진선택법: y = w1x1 + w2x2 + ....
2. 후진제거법: 전진과 반대로 하나씩 변수를 제거
3. 단계별 선택법: 전진선택법 + 후진선택법으로 변수를 추가할 때 벌벌점을 고려함
   - AIC: 편향과 분산이 최적화되는 지점 탐색, 자료가 많을수록 부정확
   - BIC: AIC 보완했으며 AIC보다 큰 패널티를 가지는 단점. 변수가 적을 때 적합

다변량 분석

상관분석
- 두 변수간의 선형적 관계가 존재하는 상태로 파악하는 분석
피어슨: 양적척도, 연속형 변수, 선형관계 크기 측정
스피어만: 서열척도, 순서형 변수, 선형/<strong>비선형적 관계</strong>
다중공선성: 다중회귀분석에서 설명변수들 사이에 상관관계가 클 때 모델을 불안정하게 만듦

다차원 척도법(MDS)
- 데이터 간의 근접성을 시각화
데이터 축소 목적, stress값이 0에 가까울수록 좋음, x/y축 해석 불가
종류 : 계량적: 양적척도, 비계량: 순서척도

<strong>주성분 분석(PCA)</strong>
- 차원축소하여 새로운 변수 생성
- 분산이 가장 큰 축이 첫번째 주성분
- 70~90%의 설명력을 갖는 수를 결정
근데 주성분의 개수를 어떻게 설정할지 알아야하므로 방법을 찾아보면 됨

1. 스크리플롯: 주성분 개수를 선택하는데 도움이 되는 그래프, 수평을 이루기 바로 전단계 개수로 선택
ex) 10, 5, 2, 2, 1, 0.5
위와 같이 구성되면 10, 과 5를 채택하여 주성분 개수를 2개로 선택
2. 바이플롯: 데이터간 유사도를 한번에 볼 수 있는 그래프, PC와 평행할수록 해당 PC에 큰 영향, 화살표의 길이가 길수록 분산이 큼
(바이플롯은 요즘 잘 안나옴)

시계열 예측
- 시간의 흐름에 따라 관찰된 자료의 특성을 파악하여 미래 예측(주가, 기온, 강수량 등등)
정상성
- 시계열 예측을 위해 모든 시점에 일정한 평균과 분산을 가지는 정상성을 만족해야 함.
- 정상시계열로 변환하는 방법 - 차분:현 시점의 자료를 이전 값으로 빼는 방법, 지수변환, 로그변환

백색잡음
- 시계열 모형의 오차항을 의미하며 원인은 알려져 있지 않음
- 평균이 0이면 가우시안 백색잡음
