<h1>데이터 이해</h1>


데이터 : 가공되지 않은 있는 그대로의 객관적 사실
정보 : 데이터로부터 가공된 자료(베스트셀러)

정성적 : 자료 특징을 풀어서 설명 - 문자, 언어(기상특보, 주관식 설문 응답)
정량적 : 자료를 수치화 - 수치, 기호 (온도, 풍속)

정형데이터 : 정보 형태가 정해짐(관계형DB, 엑셀, csv)
<strong>반정형데이터 : 뎅터를 설명하는 메타데이터를 포함(로그, html, xml, json) </strong>
비정형데이터 : 형태가 정해지지 않음(sns, 영상, 음악)

<strong>
암묵지 : 개인에게 습득되고 겉으로 드러나지 않음
형식지: 문서, 메뉴얼 등의 형상화된 지식 
</strong>
공통, 표출, 연결, 내면

DIKW 피라미드

Information(정보) : 데이터를 통해 패턴 인식
지식(knowledge) : 패턴통해 예측
지혜(wisdom) : 창의적 산물

데이터베이스 정의
DB : 일정 구조에 맞게 조직화된 데이터 집합
- 스키마 : DB 구조와 제약조건에 관한 전반적 명세
- 인스턴스 : 정의된 스키마에 따라 저장된 값
DBMS: DB를 관리
관계형 DBMS : mssql, mysql, maria DB,...
객체지향 DBMS : 정보를 객체형태로 처리
<strong>NoSQL DBMS : 비정형 데이터 저장 및 처리(Mongo DB,...)</strong>
SQL : 데이터 베이스에 접근할 수 있는 하부언어

<strong>

결측값과 이상값 검색
EDA(탐색적 자료 분석)
통계 시각화를 통해 파악
('저잔재현')

결측값
- 존재하지 않는 데이터, null/NA로 표시. 의미 있을 수 있음
- 단순대치법: 결측값 가지는 데이터 삭제
예시로 complete cases 함수로 False일 경우(결측값) 제거
- 평균대치법
- 단순확률 대치법(KNN)
- 다중 대치법: 여러번 대치(대치-분석-결합)

이상값
- 극단적인 값. 의미 있을 수도 있음(체중3kg)
- 이상값을 항상 제거하지는 않음
1. ESD: 평균으로부터 표준편차의 3배 넘는 데이터는 이상값으로 판단
2. 사분위수(Q1, Q1.5 이런식임...)

통계분석
통계학 개론
전수조사와 표본조사
- 전수조사: 전체 조사. 시간 비용이 큰 소모
- 표본 조사: 일부만 추출하여 모집단 분석

표본추출 방법
1. 랜덤 추출
2. 계통 추출: 번호 부여하여 일정 간격으로 추출
3. 집락 추출법(아파트가 있으면 1동만 추출하는 방식)
4. 층화 추출법=비례층화 추출법 -> 1학년 1명, 2학년 2명,3학년 3명,...
5. 복원/비복원 추출

자료의 척도
1. 질적 척도- 명목(대학교, 성별), 순서(학년, 순위)
2. 양적 척도- 등간척도(덧셈, 뺄셈만 가능), 비율척도(사칙연산이 가능)

첨도와 왜도
첨도: 자료의 분포가 얼마나 뾰족한지 나타내는 척도(최대값이 높게 솟아있는 것을 보는 것과 같음)
3을 기준으로 정규분포 형태를 판단함

왜도: 자료 분포의 비대칭 정도(0일 때 대칭)
<strong>왜도 < 0: 최빈값 > 중강값 > 평균값
왜도 > 0: 최빈값 < 중강값 < 평균값</strong>
