```python
url = 'https://www.melon.com/chart/index.htm'
wd.get(url)
```

```python
html = wd.page_source
h = BeautifulSoup(html, 'html.parser')
h
```

```python
h.select('tr td div.wrap input')
# h.select('#conts_section tr td div.wrap a')
```

```python
songs=h.select("tr")
songs = songs[1:]
songs[1].select('div.ellipsis a')[0].string
songs[1].select('div.ellipsis.rank01 > span > a')[0].string
songs[1].select('div.ellipsis.rank01 > span > a')[0].text
songs[1].select('div.ellipsis.rank02 > span > a')[0].text
```

```python
for song in songs:
    print(song.select('div.ellipsis.rank01 > span > a')[0].text)
```

```python
song_data = []
rank = 1
for song in songs:
    title = song.select("div.ellipsis.rank01 > span > a")[0].text
    singer = song.select("div.ellipsis.rank02 > span > a")[0].text
    song_data.append(['Melon', rank, title, singer])
    print(rank, title, singer, sep=' | ')
    rank+=1
```

```python
import pandas as pd
df = pd.DataFrame(song_data, columns=['서비스', '순위', '제목', '가수명'])
df.to_excel('melon_chart100.xlsx',index=False)
```
