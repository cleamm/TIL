# 틱택토

```python
first = input('누구먼저 시작합니까?: computer(c) or user(u) ')
tictacto = [['.','.','.'],['.','.','.'],['.','.','.',]]

import random
print('시작합니다!')
if(first == 'c'):
    row,col = random.randint(0,2), random.randint(0,2)
    tictacto[row][col] = 'o'
    for i in tictacto:
        print(i)
else:
    for i in tictacto:
        print(i)
        
print()

def check_victory1():
    com = 0
    us = 0
    for i in range(3):
        if(tictacto[i][i] == 'o'):
            com += 1
        elif(tictacto[i][i] == 'x'):
            us += 1
    if(com == 3):
        return 'com'
    elif(us == 3):
        return 'us'

def check_victory2():
    com = 0
    us = 0
    j = 2
    for i in range(3):
        if(tictacto[i][j] == 'o'):
            com += 1
        elif(tictacto[i][j] == 'x'):
            us += 1
        j -= 1
    if(com == 3):
        return 'com'
    elif(us == 3):
        return 'us'
    
def check_victory3():
    com = 0
    us = 0
    for i in range(3): # 행의 체크가 같은지 확인 후 반환
        for j in range(3): 
            if(tictacto[i][j] == 'o'):
                com += 1
            elif(tictacto[i][j] == 'x'):
                us += 1
        if com == 3:
            return 'com'
        elif us == 3:
            return 'us'
        com = 0
        us = 0

    com = 0
    us = 0
    for i in range(3): # 열의 체크가 같은지 확인 후 반환
        for j in range(3): 
            if(tictacto[j][i] == 'o'):
                com += 1
            elif(tictacto[j][i] == 'x'):
                us += 1
        if com == 3:
            return 'com'
        elif us == 3:
            return 'us'
        com = 0
        us = 0

def result():
    res = check_victory1()
    if(res == 'com'):
        return res
    elif(res == 'us'):
        return res
    res = check_victory2()
    if(res == 'com'):
        return res
    elif(res == 'us'):
        return res
    res = check_victory3()
    if(res == 'com'):
        return res
    elif(res == 'us'):
        return res
    
while True:
    print('당신의 차례입니다.')
    while True:
        r, c = map(int, input('행, 열:').split())
        if(tictacto[r][c] == 'x' or tictacto[r][c] == 'o'):
            print('이미 지정된 곳입니다. 다시 지정해주세요.')
        else:
            tictacto[r][c] = 'x'
            for i in tictacto:
                print(i)
            break
    
    result_victory = result()
    if(result_victory == 'com'):
        print('컴퓨터의 승리입니다.')
        break
    elif(result_victory == 'us'):
        print('당신의 승리입니다.')
        break

    print('컴퓨터 차례입니다.')
    while True:
        row,col = random.randint(0,2), random.randint(0,2)
        if(tictacto[row][col] != 'x' and tictacto[row][col] != 'o'):
            tictacto[row][col] = 'o'
            for i in tictacto:
                print(i)
            break

    result_victory = result()
    if(result_victory == 'com'):
        print('컴퓨터의 승리입니다.')
        break
    elif(result_victory == 'us'):
        print('당신의 승리입니다.')
        break
```

# 토익 문제

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re
url = 'https://studyingengineer.tistory.com/31'
wd = webdriver.Chrome()
wd.get(url)
html = wd.page_source
soup = BeautifulSoup(html, 'html.parser')
soup_data = soup.select('tbody tr td') # 블로그에 존재하는 테이블의 토익 단어 데이터 저장
```

```python
soup_data[2].text # 확인용
```

```python
words = [] # 영단어
means = [] # 영단어 뜻
for i in range(len(soup_data)):
    try:
        words.append(re.search('[a-z]+', soup_data[i].text)[0])
        means.append(re.search('[^a-z]+', soup_data[i].text)[0])
    except:
        pass

for i in range(5):
    print(words[i], means[i])
toeic = dict(zip(words, means))
```

```python
wrong_words = []
used_words = []
print('토익 영단어 문제풀이를 시작합니다!')
import random

while True:
    question_type = int(input('토익 문제풀이 방법을 설정해주세요 1: 영단어 입력 2:뜻 맞추기 3: 오답 노트 보기'))
    right_count = 0
    wrong_count = 0
    if question_type == 1:
        while True:
            num = random.randint(0,len(means))
            if means[num] not in used_words: # 뜻 맞추기
                used_words.append(means[num])
                answer = [i for i, value in toeic.items() if  value == means[num]]
                your_answer = input(f'{means[num]} 정답 (종료를 원하면 "exit" 입력):')
                if your_answer == 'exit':
                    break
                if your_answer in answer:
                    right_count += 1
                    print('정답입니다!')
                else:
                    wrong_count += 1
                    wrong_words.append([words[num], means[num]])
                    print('틀렸습니다!')
                    print('당신의 정답:', your_answer,'정답:', answer)

        f = open('wrong_means.txt', 'w', encoding='utf-8') # 오답 기록
        for i in wrong_words:
            total = i[0] + ',' + i[-1] + '\n'
            f.write(total)
        f.close()
        print('맞은 개수: ',right_count)
        print('틀린 개수: ',wrong_count)

    elif question_type == 2:
        while True:
            num = random.randint(0,len(words))
            if words[num] not in used_words: # 단어 적기
                used_words.append(words[num])
                your_answer = input(f'{words[num]} 정답:')
                if your_answer == 'exit':
                    break
                if toeic.get(words[num]) == your_answer:
                    right_count += 1
                    print('정답입니다!')
                else:
                    wrong_count += 1
                    wrong_words.append([words[num], means[num]])
                    print('틀렸습니다!')
                    print('당신의 정답:', your_answer,'정답:',toeic[words[num]])
        f = open('wrong_words.txt', 'w', encoding='utf-8') # 오답 기록
        for i in wrong_words:
            total = i[0] + ',' + i[-1] + '\n'
            f.write(total)
        f.close()
        print('맞은 개수: ',right_count)
        print('틀린 개수: ',wrong_count)

    elif question_type == 3:
        type2 = int(input('1: 뜻맞추기, 2: 단어맞추기'))
        if(type2 == 1):
            f = open('wrong_words.txt', 'r', encoding='utf-8') # 오답 기록
            for i in f:
                print(i)
            f.close()
        elif(type2 == 2):
            f = open('wrong_means.txt', 'r', encoding='utf-8') # 오답 기록
            for i in f:
                print(i)
            f.close()
```
