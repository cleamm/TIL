- 원칙 : 가볍지만 충분하도록

애자일 방법론
XP(익스트림 프로그래밍)
- 의사소통 개선 및 즉각적 피드백
- 5가지 가치
  - 용기
  - 단순성
  - 의사소통
  - 피드백
  - 존중
스크럼(럭비에서 나온 용어라고 함;;)
- 프로젝트 관리를 위한 상호, 점진적 개발방법론
- xp와 달리 진행 체계 수립, 역할, 정의에 중점
- 기존 폭포수 모델이나 프로토타이핑같은 모델과 달리 모든 라이프사이클을 담지 않는다.
- 5가지 가치, 확약, 전념, 정직, 존중, 용기

스크럼
PO(product owner) : 제품 책임자, 제품에 대한 요구사항 즉, 백로그를 작성하는 주체
백로그 : 요구사항 리스트
SM(scrum master) : 스크럼팀의 스크럼이 잘 수행되도록 도와주는 역할
스프린트: 짧은 기간(1~4주) 동안 사용자에게 제공하면서 피드백을 받음

스토리보드: 순차적으로 어떤 절차를 거쳐서 완성돼야 하는지 한눈에 알 수 있도록 만들어 놓는 설계도(실제구현x)
목업: 시각적인 요소로 그려놓은 틀이 잡힌 디자인의 설계도(실제구현x)
프로토타입: 실제 구현된 것처럼 테스트가 가능한 동적인 형태의 모형
유스케이스: 사용자는 공간 밖에 배치, 각 요소가 어떻게 상호작용하는지 보여주는 설계도

정형기술 검토(FTR = Formal Technical Review)
- 검토될 제품에 대한 체크 리스트를 개발
- 자원과 시간 일정을 할당
- 문제 영역을 명확히 표현하고 의제를 제한
- 제품의 검토만 집중
- 검토의 과정과 결과를 재검토
- 논쟁과 반박을 제한
- 참가자 수를 제한
- 사전 준비를 강요하고 사전에 작성한 메모를 공유
- 모든 검토자들을 위해 의미있는 훈련을 진행
- 해결책이나 개선책에 대해서 논하지 않음
그러니 거의 제한을 하지만 
제품의 검토를 집중 및 체크 리스트 개발, 해결책이나 개선책을 논하지는 않으며 의미있는 훈련을 진행함

소프트웨어 품질 관련 국제 표준
- ISO/IEC 9126 : 품질 특성과 측정 기준 제시, 평가방법이 없음
- ISO/IEC 25000 : 소프트웨어 품질평가 모델인 ISO/IEC 9126과 소프트웨어 평가절차 모델 ISO/IEC 14598, ISO/IEC12119를 통합
ㄴ 품질 관리(2500n), 모델(2501n), 측정(2502n), 요구사항(2503n), 평가(2504n)
