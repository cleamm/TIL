스크럼

정형기술 검토(FTR = Formal Technical Review)
- 검토될 제품에 대한 체크 리스트를 개발
- 자원과 시간 일정을 할당
- 문제 영역을 명확히 표현하고 의제를 제한
- 제품의 검토만 집중
- 검토의 과정과 결과를 재검토
- 논쟁과 반박을 제한
- 참가자 수를 제한
- 사전 준비를 강요하고 사전에 작성한 메모를 공유
- 모든 검토자들을 위해 의미있는 훈련을 진행
- 해결책이나 개선책에 대해서 논하지 않음
그러니 거의 제한을 하지만 
제품의 검토를 집중 및 체크 리스트 개발, 해결책이나 개선책을 논하지는 않으며 의미있는 훈련을 진행함

소프트웨어 품질 관련 국제 표준
- ISO/IEC 9126 : 품질 특성과 측정 기준 제시, 평가방법이 없음
- ISO/IEC 25000 : 소프트웨어 품질평가 모델인 ISO/IEC 9126과 소프트웨어 평가절차 모델 ISO/IEC 14598, ISO/IEC12119를 통합
ㄴ 품질 관리(2500n), 모델(2501n), 측정(2502n), 요구사항(2503n), 평가(2504n)

리눅스 UMASK
- 모든 파일과 디렉토리 - 소유권, 권한 존재(root를 의미하는 듯)
- 소유권 - 소유자, 소유그룹, 기타 사용자(윈도우도 마찬가지긴 함)
- 권한 - 읽기, 쓰기, 실행(execute)
- UMASK - 소유권에 대한 권한을 지정해주는것, 권한 허가x, 권한 제한O
- 파일의 최대 권한은 666, 디렉토리는 777
읽기  쓰기  실행
r       w   x
4       2   1  총합 7로 파일은 실행이 없고, 디렉토리는 실행까지 가능함
2^2   2^1   2^0  => 2의 제곱단위로 변하는 것임
표기로는 읽기 쓰기가 되면 'rw-' 로 표기
읽기만 되면 'r--' 전부 되면 rwx로 표기

디렉토리로는
소유자 소유그룹 기타사용자
6       6       0
rw-     rw-     ---(없으면 대시3개)

예제: 리눅스에서 생성된 파일 권한이 644일 경우 umask 값은? 정답: 666임(왜냐면 파일 권한에 관한 내용이기 때문.)

공유 최대를 위한 병행 제어에서의 locking
- 한번에 로킹(locking)할 수 있는 단위를 '로킹 단위'라고 함
- db,파일,레코드 등은 로킹 단위가 될 수 있음
- 로킹 단위가 작아지면 로킹 오버헤드가 증가함
- locking이 크면 제어는 간단하지만 병행성이 감소
- locking이 작으면 제어가 어려우나 병행성이 증가
