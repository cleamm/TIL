```python
import pandas as pd
import numpy as np
 
data = {
    'Gender' : ['Male', 'Male', 'Female', 'Female', 'Male','Male','Male','Male','Female', 'Female'],
    'Fruit' : ['Apple',np.nan,'Banana','Apple',np.nan,'Strawberry','Banana','Banana','Apple', np.nan],
    'Age' : [20, 14, 13, 22, 14, 32, 31, 22,23, 33],
    'Height' : [187,np.nan,155,165,177,np.nan,170,179,np.nan, 166]
}
df = pd.DataFrame(data); df
```

```python
# 1) 범주형 결측값 대체 방법 : 성별로 그룹화 -> 빈도수가 가장 높은 값으로 대체
# 2) 연속형 결측값 대체 방법 : 성별로 그룹화 -> 통계값(평균, 중위수, ..)으로 대체
```

```python
df # Fruit 열의 결측값 대체 : 
# gender가 male인 경우에는 male그룹에 대한 fruit열 값의 최빈값으로 대체
# gender가 female인 경우에는 female그룹에 대한 fruit열 값의 최빈값으로 대체
```

```python
df.Gender.value_counts()
# Male      6
# Female    4
df.Gender.value_counts().idxmax()
```

```python
# df의 fruit열에 대해 결측값을 그룹별 통계값으로 대체
df.Fruit.fillna(df.groupby('Gender')['Fruit'].transform(lambda x: x.value_counts().idxmax()), inplace=True)
# x에는 Gender별 Fruit열이 전달됨
df
```

```python
# height열의 NaN값을 각 성별에 따른 평균값으로 수정하기
df.Height.fillna(df.groupby('Gender')['Height'].transform('mean'), inplace=True)
# df.Height.fillna(df.groupby('Gender')['Height'].transform(lambda x: x.mean()))
# df
```

SVM(서포트 벡터 머신): 데이터 분류를 위해 마진이 최대가 되는 결정 경계선(decision boundary)을 찾아내는 머신러닝 방법

SVM을 이용해서 이진 분류를 하는 것은 최적(optimal = 일반화가 잘된 모델)의 결정 경계선을 찾는 것이 중요함

왜 하나하나 다 맞게 만들지 않는가? = 그것은 과적합이기 때문에 일반화가 중요해짐

그리고 단순히 결정경계를 찾는 것이 아닌 '최적의' 결정 경계를 찾아야 함

잘 나눈 선은 선을 수평으로 뒀을 때 점과의 수직거리가 평균적으로 먼것

1. 결정 경계선
- 결정 경계는 N차원 - 1차원임(그릠 참고할 것)
- 입체에서는 면으로 구분, 면에서는 선으로 구분하게 됨
2. 서포트 벡터
- 벡터는 2차원 공간 상에 나타난 데이터 포인트
3. 마진
- 결정 경계 사이의 거리가 최적화 되어야 함
4. 비용
- 얼마나 틀린 분류를 하게 되는 것을 허용하는지에 대한 결정(그림 참고)
5. 커널 트릭
- 차원을 변형시킴
- 차원을 변형시킬 땐 좌표값이 변형되는데 부족한 좌표값은 기존 좌표값의 제곱을 하여 그 값으로 적용

svm
- 특성이 다양한 데이터 분류에 강함
- 파라미터를 조정하여 과대/과소적합 대응 가능
- 적은 학습 데이터로도 정확도가 높은 분류 성능
- 데이터 전처리가 중요
- 특성이 많으면 결정 경계 및 데이터 시각화가 어려움

```python
# 최적의 svm 파라미터 찾기
# 그리드 서치를 사용하여 최적 비용 및 감마 알아내기
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
# 일일이 테스트해서 어느쪽이 좋은지 테스트는 할 수 있지만 한번에 탐색할 수 있는 방법이 있다면 그 방법을 적용하는 것이 좋으므로 그리드 서치를 사용
```

```python
# 타이타닉을 이용한 SVM모델 적용
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.svm import SVC
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('titanic/train.csv')
test = pd.read_csv('titanic/test.csv')
train.head()
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
survived = train[train['Survived'] == 1]['Sex'].value_counts()
dead = train[train['Survived'] == 0]['Sex'].value_counts()
df = pd.DataFrame([survived, dead])
df.index = ['Survived', 'Dead']
df.head()
```

```python
train_test_data = [train, test]
train_test_data[1]
```

```python
'Kelly, Mr. James'

for dataset in train_test_data:
    dataset['Title'] = dataset['Name'].str.extract('([A-Za-z]+)\.')
train.head() # title열이 추가됨
```

```python
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 2,
                 "Master": 3, "Dr": 3, "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3,"Countess": 3,
                 "Ms": 3, "Lady": 3, "Jonkheer": 3, "Don": 3, "Dona" : 3, "Mme": 3,"Capt": 3,"Sir": 3 }
for dataset in train_test_data:
    dataset['Title'] = dataset['Title'].map(title_mapping)
train.head()
```

```python
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
```

```python
sex_mapping = {"male": 0, "female": 1}
for dataset in train_test_data:
    dataset['Sex'] = dataset['Sex'].map(sex_mapping)
train.head()
```

```python
train["Age"].fillna(train.groupby("Title")["Age"].transform("median"), inplace=True)
test["Age"].fillna(test.groupby("Title")["Age"].transform("median"), inplace=True)
train.head()
```

```python
for dataset in train_test_data:
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 62, 'Age'] = 4
train.head()
```

```python
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
train.head()
```

```python
embarked_mapping = {"S": 0, "C": 1, "Q": 2}
for dataset in train_test_data:
    dataset['Embarked'] = dataset['Embarked'].map(embarked_mapping)
train.head()
```

```python
test.loc[test['Fare'].isnull(), 'Fare'] = test[(test['Pclass'] == 3) & (test['Embarked'] == 0)]['Fare'].median()
```

```python
for dataset in train_test_data:
    dataset.loc[ dataset['Fare'] <= 17, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 17) & (dataset['Fare'] <= 30), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 30) & (dataset['Fare'] <= 100), 'Fare'] = 2
    dataset.loc[ dataset['Fare'] > 100, 'Fare'] = 3
train.head()
```

```python
for dataset in train_test_data:
    dataset['Cabin'] = dataset['Cabin'].str[0]
train.head()
```

```python
cabin_mapping = {"A": 0, "B": 0.4, "C": 0.8, "D": 1.2, "E": 1.6, "F": 2, "G": 2.4, "T": 2.8}
for dataset in train_test_data:
    dataset['Cabin'] = dataset['Cabin'].map(cabin_mapping)
train.head()
```

```python
train["Cabin"].fillna(train.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
test["Cabin"].fillna(test.groupby("Pclass")["Cabin"].transform("median"), inplace=True)
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
train.head()
```

```python
family_mapping = {1: 0, 2: 0.4, 3: 0.8, 4: 1.2, 5: 1.6, 6: 2, 7: 2.4, 8: 2.8, 9: 3.2, 10: 3.6, 11: 4}
for dataset in train_test_data:
    dataset['FamilySize'] = dataset['FamilySize'].map(family_mapping)
train.head()
```

```python
features_drop = ['Ticket', 'SibSp', 'Parch']
train = train.drop(features_drop, axis=1)
test = test.drop(features_drop, axis=1)
```

```python
train = train.drop(['PassengerId'], axis=1)
```

```python
train_data = train.drop('Survived', axis=1)
target = train['Survived']
```

```python
test_data = test.drop("PassengerId", axis=1).copy()
```

```python
test_data.head()
```

```python
# svm으로 모델링
from sklearn.svm import SVC 
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC
```

```python
clf = SVC() # 기본 rbf, c=1.0, gamma='scale' scale = 1/(열의 개수*분산)
clf.fit(train_data, target)
```

```python
answer = test.Pos
test.drop('Pos', axis=1, inplace=True)
test.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=0) # 교차검증
for i in range(1, 20, 2):
    clf = KNeighborsClassifier(n_neighbors=i) # 모델 생성
    score = cross_val_score(clf, train_data, target, cv=kfold, n_jobs=1, scoring='accuracy')
    clf.fit(train_data, target) # 모델 학습
    prediction = clf.predict(test)
    print((answer == prediction).sum()/len(answer)) # 정확도
```

1일때가 정확도가 가장 높음
