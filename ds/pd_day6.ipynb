```python
#1. df1과 df2에 대해 merge 연습, on, left_on, right_on, how
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
df1, df2
```

```python
pd.merge(df1, df2, on='key') # df1 - abc, df2 - abd 중복되는 값은 ab이므로 3*2 총 6행
pd.merge(df1,df2, how='left') # 왼쪽 기준 병합
pd.merge(df1,df2, how='right') # 오른쪽 기준 
pd.merge(df2, df1, on='key') # 기본은 먼저 입력된 데이터프레임 기준으로 사용됨
pd.merge(df2,df1, left_on='key', right_on='key') # 데이터는 같고 컬럼명이 다르면 사용할 것. 결과는 위와 같음
df3 = pd.merge(df1, df2, left_on='data1', right_on='data2')
df3.rename(columns={'key_x':'df1', 'key_y': 'df2'}, inplace=True)
df3
```

```python
#2. left과 right에 대해 merge 연습, on, left_on, right_on, how
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
left, right
```

```python
pd.merge(left, right) # 내부적으로 있는 데이터만 출력
pd.merge(left, right, how='outer') # 총 데이터를 출력
pd.merge(left, right, how='left') # left 데이터프레임의 데이터를 기준으로 병합
pd.merge(left, right, on=['key1', 'key2']) # pd.merge(left, right)와 결과가 같음 = 2개의 열을 기준으로 했는데 중복열도 2개이므로
pd.merge(left, right, on='key2') # key2를 기준으로 병합
pd.merge(left, right, on='key1') # key1을 기준으로 병합
pd.merge(left, right, on='key1', how='right') # key1의 병합 후 이후의 내용은 right를 기준으로 병합
pd.merge(left, right, left_index=True, right_index=True) # 중복되는 인덱스값을 기준으로 병합
```

```python
#3. merge 열 이름과 행 인덱스를 기준으로 병합 연습
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                      "value": pd.Series(range(6), dtype="Int64")})
right1 = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
left1, right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True) # left1의 열과 right1의 인덱스를 기준으로 병합
right2 = right1.reset_index() # 인덱스를 열로 변경 후 rigth2로 초기화
right2.rename(columns={'index':'key'}, inplace=True) # 컬럼명 변경
pd.merge(left1, right2)
```

```python
#4. merge연습, index 기준
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=["a", "c", "e"],
                     columns=["Ohio", "Nevada"]).astype("Int64")
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=["b", "c", "d", "e"],
                      columns=["Missouri", "Alabama"]).astype("Int64")
left2, right2
```

```python
pd.merge(left2, right2, left_index=True, right_index=True)
pd.merge(left2, right2, left_index=True, right_index=True, how='outer')
left2.reset_index(inplace=True)
right2.reset_index(inplace=True)
df = pd.merge(left2, right2)
df.rename(columns={'index':'eng'}, inplace=True)
df
pd.merge(left2, right2, how='outer') # 모든 조합 출력
```

```python
#5. concat 연습, axis, join 속성 값 변경 설정
s1 = pd.Series([0, 1], index=["a", "b"], dtype="Int64")
s2 = pd.Series([2, 3, 4], index=["c", "d", "e"], dtype="Int64")
s3 = pd.Series([5, 6], index=["f", "g"], dtype="Int64")
s1, s2, s2
```

```python
pd.concat([s1,s2,s3]) # np.vstack과 같은 효과를 보임
pd.concat([s1,s2,s3], axis=1)
s12 = pd.concat([s1,s2], axis=1)
s2.name = 's2' # name이 있어야 하므로 추가
s3.name = 's3'
s12.join(s3, how='right') # 병합후 s3를 기준으로 출력
s12.join(s3, how='outer')
pd.DataFrame(s1).join(s2, how='outer') # 데이터프레임에 join이 있음
pd.DataFrame(s1).join([s2,s3], how='outer') # 여러개의 시리즈를 데이터프레임에 추가
```

```python
'''
- survived열 값으로 전체 생존율 출력
- pclass열 값이 1인 자료의 개수를 출력
- age가 10보다 작은 승객에 대한 생존율
- embarked(항구의 종류임) 중에서 각 항구별 승선인원 출력(embark_town에 나와있음)
'''
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
```
