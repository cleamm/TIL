```python
# 데이터마이닝 - 연관규칙(association rules)
# https://velog.io/@sohtks/Recommendation-System-Association-Rule-Mining-%EC%97%B0%EA%B4%80-%EA%B7%9C%EC%B9%99
# https://myeonghak.github.io/recommender%20systems/RecSys-%EC%97%B0%EA%B4%80-%EA%B7%9C%EC%B9%99-%EC%B6%94%EC%B2%9C-(Association-Rule)/
# 블로그에 관련 설명이 있음
# 추가로 대학원에서 진행하는 내용이라고 하심(?난 왜 3학년 때 배움)
# 동시 구매가 예상되는 아이템을 찾을 수 있는 규칙을 찾는 것
# 기저귀를 산 사람이 맥주를 사더라(?)
# 우유와 빵을 사는 사람은 달걀과 콜라를 사더라
# 맥주와 빵을 사는 사람은 우유를 사더라
# 위와 같은 규칙을 찾는 것임

# 위와 같은 규칙을 발견하면 어떻게 하면 좋을까? = 비슷한 코너에 배치, 진열대 옆에 배치, 세트로 판매 등등..
# 그럼 기저귀 가게가 파업했다고 가정했을 때, 영향을 받는 제품이 무엇일까? = 위와 같은 상황이었다면 맥주의 판매량도 떨어질 것으로 예측


# TID   Items
# 1     Bread, Milk
# 2     Bread, Diaper, Beer, Eggs
# 3     Milk, Diaper, Beer, Coke 
# 4     Bread, Milk, Diaper, Beer
# 5     Bread, Milk, Diaper, Coke 


# 한가지 이상의 아이템을 모은 것을 아이템셋(itemset)이라고 함 
# support는 '지지도'라고 부름. 이는 지정된 아이템들이 포함된 아이템셋이 몇개인지에 따라 표현이 바뀜.
# 하나의 아이템들을 트랜잭션이라고 하며 비율을 계산할 때는 지지도 개수가 2건, 트랜잭션이 5건이면 지지도는 2/5=0.4로 계산이 됨
# frequent itemset(빈발 항목 집합): 최소 임계값 이상(직접 설정해야됨)
# 편의점에서 물이나 담배같은 물품, 음료 등은 거의 지지도가 0.1이상은 됨
# 연관규칙은 일정 거래 이상이 되는 물품에 대해서 적용하는 것임. 벌초물품같은 경우는 자주 사용하지 않으므로 연관규칙을 적용하지 않음
# 반대로 담배살때 라이터도 사더라! 하는 것은 무시해야됨.
# 결국 데이터 마이닝은 인사이트(insight)를 얻기 위해서 하는 것

# X -> Y
# {우유, 기저귀}(X임) -> {맥주}(Y임)
# 위와 같은 케이스를 보자면 우유, 기저귀, 맥주를 모두 포함하는 지지도를 구하는 것으로 s = 우유, 기저귀 & 맥주 포함누적계산 = 2/5
# 수식으로는 S(X->Y)라고 할 수 있음
# C(X->Y)도 있는데 이는 '신뢰도'(Confidence)라고 함
# x가 있는 리스트에서 Y도 있는 경우를 계산하는 것임
# 위 아이템 셋을 봤을 때는 TID 3, 4, 5가 X에 해당하며 4,5가 Y도 포함되어 있는 경우로, 2/3(66.67%)으로 계산됨
# 연관분석에서는 순서를 안봄(순서가 저장되지 않음)

# 트랜잭션 T가 주어졌을 때 목표는 다음과 같음
# 지지도 >= 최소 지지도 임계값
# 신뢰도 >= 최소 신뢰도 임계값

# 서로 다른 종류의 아이템을 가진 5개 항목이 있을 때 생성 가능한 모든 조합(1개~5개 묶인 조합까지)은 32개임
# 이에 대해서 계산 복잡도를 계산하는 방법은 다음과 같음
# 3^d - 2^d+1 + 1 => 3**5 - 2**6 + 1 이렇게 된다고 볼 수 있음(총 180가지 규칙이 있음)
# 만약 1000가지 품목에 대해서 규칙을 보려고 한다면 d에 1000을 대입하면 되는데...컴퓨터로 할 수 없는 영역임...(단순연산이면 모르겠지만)

# cde가 빈발항목이라면 minsup(최소지지도) = 0.1(직접 설정)보다 크면 frequent itemset임
# 예시로 {물,우유,라면}이 빈발아이템이라면 minsup보다 크거나 같다면 물이라는 항목만 두고 봤을 때는 0.1보다 큰가? = 당연히 맞음
# 혹여나 최소지지도보다 작은 항목들이라면 Pruned supersets(가지치기)를 하여 계산하지 않으면 됨
# 예시로 abcde가 있는데 ace가 0.1을 충족하지 못한다면, a,c,e가 포함된 모든 항목을 제외하고 계산하면 됨
# 위와 같은 것이 apriori라고 함
# 물론 위의 기능을 하는 모델이 있으므로 최소 지지도를 지정해주면 알아서 해결해줌
# 이런 모델은 보통 '내가 우유를 사고 뭘 사려고 했더라?' 할때 서포트 해주는 것임

# 궁극적으로는 lift(향상도)를 찾기 위해서 위의 내용을 배우고 구한 것임
# lift = X -> Y / Y == P(X|Y)/P(Y) => x:기저귀, y:맥주일때, P(기저귀|맥주)(맥주를 샀을때 기저귀도 사는 비율)/P(기저귀)(기저귀를 산 비율)
# 총 100건, 맥주를 10건, 기저귀 5건, 맥주를 샀을 때 기저귀 2건이라면 다음과 같이 계산함
# (2/10)/(5/100) => 0.2/0.05 => 4
# 일반 고객들이 기저귀를 구매한 비율 대비 맥주를 구매한 고객들 중에 기저귀를 구매한 비율에 대한 비율 을 의미함
# 만약 값이 1이라면? 관계가 거의 없는 것임
# 위는 4라는 결과가 나왔는데 4라면 연관성이 매우 높다!라고 할 수 있음.(일반사람 대비 4배나 비율이 높다는 것임)
# 정확히는 1을 넘으면 관련이 있다고 보기는 한데, 1보다 조금 크면 유의해야함.

# 향상도의 공식을 보면 c(x->y)/s(y) = lift(x->y) 임

# 연관규칙(lift가 1보다 큰 규칙 발견)
# minConf, minSup값 설정
# Apriori 알고리즘 -> 연관 규칙 생성
# lift를 기준 내림차순 정렬 -> 연관규칙 추출
# 마케팅 단계...
# 위와 같은 수순으로 진행됨
```

# 라이브러리 가져오기|

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('GroceryStoreDataSet.csv', names=['products'], sep=',')
df # 현재 sep를 설정했음에도 구분이 안되는데 이런 경우는 열이 한개이기 때문에 데이터가 분리되지 않는 것임
```

# 데이터 전처리

```python
data = list(df.products.apply(lambda x:x.split(',')))
data # 2차원 구조로 변경해줘야 함.
# data에 있는 하나의 행이 트랜잭션임
```

# 트랜잭션인코더로 인코딩

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
trans_data = te.fit(data).transform(data)
trans_data # n번째 트랜잭션에서 해당 아이템이 있을 경우 True를 의미함
# te.columns_를 사용하면 해당 True값의 데이터가 무엇인지 알 수 있음
```

```python
te.columns_
```

# 컬럼과 데이터 병합

```python
df = pd.DataFrame(trans_data, columns=te.columns_)
df
```

# Apriori
지지도와 신뢰도를 설정해야됨. 여기서는 지지도 0.2, 신뢰도 0.6으로 할 것임

```python
# 지지도 계산
sp = apriori(df, min_support=0.2, use_colnames=True, verbose=True).sort_values('support', ascending=False)
sp
```

```python
# 신뢰도 계산
df_ar = association_rules(sp, metric='confidence', min_threshold=0.6)
df_ar
```
